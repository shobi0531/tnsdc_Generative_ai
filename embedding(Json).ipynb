```python
import pandas as pd
import json

# Load JSON data
with open('entries .json', 'r') as file:
    data = json.load(file)

# Function to flatten JSON data
def flatten_json(json_data, parent_key='', sep='_'):
    items = {}
    for key, value in json_data.items():
        new_key = parent_key + sep + key if parent_key else key
        if isinstance(value, dict):
            items.update(flatten_json(value, new_key, sep=sep))
        elif isinstance(value, list):
            for i, v in enumerate(value):
                items.update(flatten_json(v, new_key + sep + str(i), sep=sep))
        else:
            items[new_key] = value
    return items

# Flatten JSON data
flattened_data = [flatten_json(entry) for entry in data]

# Create DataFrame
df = pd.DataFrame(flattened_data)

# Display DataFrame
print(df)
```

```python
import torch
from transformers import BertTokenizer, BertModel
from tqdm import tqdm
```

```python
# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()  # Set the model to evaluation mode
```

```python
def embed_text(text):
    # Tokenize input text
    input_ids = tokenizer.encode(text, add_special_tokens=True, return_tensors='pt')

    # Obtain BERT embeddings
    with torch.no_grad():
        outputs = model(input_ids)
        embeddings = outputs.last_hidden_state.mean(dim=1)  # Average pooling of embeddings

    return embeddings
```

```python
# Update your summarize_condition function to embed the summaries
def summarize_condition_with_embedding(condition):
    summary = ''
    field = condition['properties']['field']
    value = condition['properties']['value']
    if condition['type'] == 'condition':
        summary += f'"{field}" field does not match the value "{value}".'
    elif condition['type'] == 'fit':
        summary += f'"{field}" field matches the value "{value}".'
    elif condition['type'] == 'condition_else':
        summary += f'"{field}" field does not match the value "{value}" (else condition).'
    else:
        summary += 'Unknown condition type.'
    # Embed the summary using BERT
    summary_embedding = embed_text(summary)
    return summary, summary_embedding
```

```python
def summarize_conditions_with_embedding(data):
    summaries = []
    embeddings = []
    for item in tqdm(data, desc="Summarizing with embedding"):
        if item['type'] == 'condition_then' or item['type'] == 'condition_else':
            children = item['children1']
            for child in children:
                summary, summary_embedding = summarize_condition_with_embedding(child)
                summaries.append(summary)
                embeddings.append(summary_embedding)
    return summaries, embeddings
```

```python
import json

def retrieve_matching_fields(json_data, value):
    matching_fields = []

    def search_fields(obj):
        if isinstance(obj, dict):
            for key, val in obj.items():
                if key == "field" and val == value:
                    matching_fields.append(obj["field"])
                elif isinstance(val, (dict, list)):
                    search_fields(val)
        elif isinstance(obj, list):
            for item in obj:
                search_fields(item)

    search_fields(json_data)
    return matching_fields

# Load JSON data from file
with open('entries .json', 'r') as file:
    data = json.load(file)
print(data)
# Value to search for
search_value = "SYS_PARAM_TIMEZONE"

# Retrieve matching fields
matching_fields = retrieve_matching_fields(data, search_value)

# Print the matching fields
print("Fields that match the value:")
for field in matching_fields:
    print(field)
```
